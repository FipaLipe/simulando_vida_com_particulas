```python
def forca(dist, r1, r2, f1, f2):

    if 0 < dist <= r1:
        return (dist * abs(f2 - f1) / r1) - abs(f1)
    
    if r1 < dist < r2:
        return f2 * abs(r2 - dist) / abs(r2 - r1)

    return 0
```

```python
r1 = 20
r2 = 60
f1 = -10
f2 = 20
```

```python
x = [i/10 for i in range (1,800)]
```

```python
y = [forca(xi, r1, r2, f1, f2) for xi in x]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y)
plt.show()
```

```python
def forca(dist, atracao, raio):

    if 0 < dist <= raio:
        return (dist / raio) - 1

    if raio < dist < 1:
        return atracao * (1 - abs(2*dist - 1 - raio) / (1 - raio))

    return 0
```

```python
a = 0.6
r = 0.3

x = [i/100 for i in range (1,120)]
y = [forca(xi, a, r) for xi in x]

plt.plot(x, y)
plt.axe
plt.show()
```
